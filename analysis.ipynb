```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_parquet("analysis.parquet")
```

```python
plt.plot(df["time_s"], df["speed_integral_kmh"])
plt.plot(df["time_s"], df["accel_integral_mss"]*100)
plt.plot(df["time_s"], df["speed_differential_kmh"])
plt.plot(df["time_s"], df["accel_differential_mss"]*100)
plt.plot(df["time_s"], df["section"])
plt.legend()
plt.show()
```
